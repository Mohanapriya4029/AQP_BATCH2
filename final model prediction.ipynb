```python
import pandas as pd
import numpy as np
import seaborn as sns
```

```python
df=pd.read_csv('city_day.csv')
df.head()
```

```python
df.reset_index(drop=True,inplace=True)
df.head()
```

```python
df.isnull().sum()
```

```python
df.info()
```

```python
df['PM2.5']=df['PM2.5'].fillna(df['PM2.5'].mean())
df['PM10']=df['PM10'].fillna(df['PM10'].mean())
df['NO']=df['NO'].fillna(df['NO'].mean())
df['NO2']=df['NO2'].fillna(df['NO2'].mean())
df['NOx']=df['NOx'].fillna(df['NOx'].mean())
df['NH3']=df['NH3'].fillna(df['NH3'].mean())
df['CO']=df['CO'].fillna(df['CO'].mean())
df['SO2']=df['SO2'].fillna(df['SO2'].mean())
df['O3']=df['O3'].fillna(df['O3'].mean())
df['Benzene']=df['Benzene'].fillna(df['Benzene'].mean())
df['Toluene']=df['Toluene'].fillna(df['Toluene'].mean())
df['Xylene']=df['Xylene'].fillna(df['Xylene'].mean())
df['AQI']=df['AQI'].fillna(df['AQI'].mode()[0])
df['AQI_Bucket']=df['AQI_Bucket'].fillna('Moderate')
```

```python
from sklearn.preprocessing import LabelEncoder

label_encoder_state = LabelEncoder()
df['City_encode'] = label_encoder_state.fit_transform(df['City'])
```

```python
df.head()
```

```python
df['Day_date'] = pd.to_datetime(df['Date'], format='%Y-%m-%d').dt.day
df['month_date'] = pd.to_datetime(df['Date'], format='%Y-%m-%d').dt.month
df['year_date'] = pd.to_datetime(df['Date'], format='%Y-%m-%d').dt.year

df.drop(['Date'], axis=1, inplace=True)

df.head()
```

```python
import seaborn as sns
import matplotlib.pyplot as plt
import pandas as pd


numerical_cols = ["PM2.5","PM10","NO","NO2","NH3","CO","O3","Benzene","Toluene","Xylene","AQI","City_encode","Day_date","month_date","year_date"]  
df_numerical = df[numerical_cols]

# Calculate the correlation matrix
corr = df_numerical.corr()

# Plot the heatmap
plt.figure(figsize=(12, 10))
sns.heatmap(corr, annot=True, cmap=plt.cm.Reds)
plt.show()
```

```python
X=df.drop(['AQI',"PM2.5","PM10","NO","NO2","NH3","CO","O3","Benzene","Toluene","Xylene","NOx","SO2","City","AQI_Bucket"],axis=1)
y=df['AQI']
print(X.shape)
print(y.shape)
```

```python
X.head()
```

```python
from sklearn.model_selection import train_test_split
X_train,X_test,y_train,y_test = train_test_split(X,y,test_size=0.2)
```

```python
from sklearn.metrics import classification_report,accuracy_score
from sklearn.ensemble import RandomForestRegressor
from sklearn.metrics import r2_score
from sklearn.metrics import mean_squared_error


rf=RandomForestRegressor()
rf.fit(X_train,y_train)
```

```python
y_pred1 = rf.predict(X_test)
accuracy1 = r2_score(y_test, y_pred1)*100
print("R2 Score:", accuracy1)
```

```python
loss1 = mean_squared_error(y_test, y_pred1)/100
print("Mean Squared Error:", loss1)
```

```python
from sklearn.ensemble import ExtraTreesRegressor
ET=ExtraTreesRegressor()
ET.fit(X_train,y_train)
```

```python
y_pred2 = ET.predict(X_test)
accuracy2= r2_score(y_test, y_pred2)*100
print("R2 Score:", accuracy2)
```

```python
loss2 = mean_squared_error(y_test, y_pred2)/100
print("Mean Squared Error:", loss2)
```

```python
from sklearn.linear_model import LinearRegression

linear_model = LinearRegression()
linear_model.fit(X_train,y_train)
```

```python
y_pred3 = linear_model.predict(X_test)
accuracy3= r2_score(y_test, y_pred3)*100
print("R2 Score:", accuracy3)
```

```python
loss3 = mean_squared_error(y_test, y_pred3)/100
print("Mean Squared Error:", loss3)
```

```python
from sklearn.preprocessing import MinMaxScaler
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import LSTM, Dense, Dropout,Flatten
from tensorflow.keras.utils import to_categorical

scaler = MinMaxScaler()
X_train_scaled = scaler.fit_transform(X_train)
X_test_scaled = scaler.transform(X_test)

X_train_reshaped = X_train_scaled.reshape((X_train_scaled.shape[0], X_train_scaled.shape[1], 1))
X_test_reshaped = X_test_scaled.reshape((X_test_scaled.shape[0], X_test_scaled.shape[1], 1))
Y_train_encoded = to_categorical(y_train)
Y_test_encoded = to_categorical(y_test)

```

```python

from keras.models import Sequential
from keras.layers import LSTM, Bidirectional, Dropout, Dense
from keras.callbacks import EarlyStopping
import os

# Assuming you have already defined X_train_reshaped, X_test_reshaped, Y_train_encoded, and Y_test_encoded

bi_lstm_model = Sequential()
bi_lstm_model.add(Bidirectional(LSTM(128, input_shape=(X_train_reshaped.shape[1], X_train_reshaped.shape[2]), kernel_regularizer='l2')))
bi_lstm_model.add(Dropout(0.5))
bi_lstm_model.add(Dense(1, activation='sigmoid'))

# compile the model
# Compiling the model
bi_lstm_model.compile(
    # Optimizer for updating the network weights
    optimizer='adam',
   
)

# fit the model with early stopping
bi_lstm_history = bi_lstm_model.fit(X_train_reshaped, Y_train_encoded, epochs=5, batch_size=32, 
                                    validation_data=(X_test_reshaped, Y_test_encoded))

#

```

```python
y_pred4 = bi_lstm_model.predict(X_test)
r2= r2_score(y_test, y_pred4)*10
accuracy4=abs(r2)
accuracy4
```
